```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://theconferencewebsite.com/")
page.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
page_info = soup.find(class_="content")
print(page_info.prettify())
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
page_info = soup.find(class_="content")
specific_info = page_info.find(class_="about")
print(specific_info.prettify())
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
page_info = soup.find(class_="content")
specific_info = page_info.find(class_="upcoming")
print(specific_info.prettify())
```
